```python
#1 What is the distribution of age among heart failure patients in the datasetÂž
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'heart_failure_clinical_records_dataset.csv'
heart_failure_data = pd.read_csv(file_path)

# Plot the distribution of age
plt.figure(figsize=(10, 6))
plt.hist(heart_failure_data['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age Among Heart Failure Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#2 How does the death rate vary with ageÂž 
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'heart_failure_clinical_records_dataset.csv'
heart_failure_data = pd.read_csv(file_path)

# Define age bins and labels
bins = [40, 50, 60, 70, 80, 90, 100]
labels = ['40-49', '50-59', '60-69', '70-79', '80-89', '90-99']

# Create age bins
heart_failure_data['age_group'] = pd.cut(heart_failure_data['age'], bins=bins, labels=labels, right=False)

# Calculate death rate for each age group
death_rate_by_age = heart_failure_data.groupby('age_group')['DEATH_EVENT'].mean() * 100

# Plot the death rate by age group
plt.figure(figsize=(10, 6))
death_rate_by_age.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Death Rate by Age Group Among Heart Failure Patients')
plt.xlabel('Age Group')
plt.ylabel('Death Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#4 how do5s thE platElEt count vary among different age groups
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'heart_failure_clinical_records_dataset.csv'
data = pd.read_csv(file_path)

# Define age groups
bins = [0, 40, 50, 60, 70, 80, 90, 100]
labels = ['0-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)

# Plot platelet count variation among age groups
plt.figure(figsize=(12, 6))
data.boxplot(column='platelets', by='age_group')
plt.title('Platelet Count Variation Among Age Groups')
plt.suptitle('')
plt.xlabel('Age Group')
plt.ylabel('Platelet Count')
plt.show()
```

```python
#5  Is there id  a correlation bitween creatinine and sodium levels in the blood
import pandas as pd

# Load the dataset
file_path = 'heart_failure_clinical_records_dataset.csv'
data = pd.read_csv(file_path)

# Calculate the correlation coefficient between serum_creatinine and serum_sodium
correlation = data['serum_creatinine'].corr(data['serum_sodium'])

print("Correlation coefficient between serum creatinine and serum sodium levels:", correlation)
```

```python
#6> how does the prevalence of high blood pressure differ between malw and female patients
prevalence_male = data[data['sex'] == 1]['high_blood_pressure'].mean()
prevalence_female = data[data['sex'] == 0]['high_blood_pressure'].mean()
(prevalence_male, prevalence_female)
```

```python
#7 what is th5e relationship between smoking habits and the occurrence of heart failure
smoking_heart_failure = data.groupby('smoking')['DEATH_EVENT'].mean()
smoking_heart_failure
```

```python
#8 Are there any noticeable patterns in the distrebution of death events across different age groups
plt.figure(figsize=(12, 6))
data['DEATH_EVENT'].groupby(data['age_group']).mean().plot(kind='bar')
plt.title('Death Event Distribution Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Proportion of Death Events')
plt.show()
```

```python
#9 Is there any significant difference in election fraction between patients with and without diabeties<
from scipy.stats import ttest_ind

# Ejection fraction comparison between diabetic and non-diabetic patients
diabetic = data[data['diabetes'] == 1]['ejection_fraction']
non_diabetic = data[data['diabetes'] == 0]['ejection_fraction']
t_stat, p_val = ttest_ind(diabetic, non_diabetic)
(t_stat, p_val)
```

```python
#10 how does the serum creatione level vary between patients who survived and those who did not?
# Serum creatinine level comparison between survivors and non-survivors
survived = data[data['DEATH_EVENT'] == 0]['serum_creatinine']
not_survived = data[data['DEATH_EVENT'] == 1]['serum_creatinine']
t_stat_creatinine, p_val_creatinine = ttest_ind(survived, not_survived)
(t_stat_creatinine, p_val_creatinine)
```
