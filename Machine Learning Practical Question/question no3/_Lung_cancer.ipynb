```python
#Load the Data
import pandas as pd

# Load the dataset
file_path = 'dataset.csv'
lung_cancer_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(lung_cancer_data.head())
```

```python
#Check the Structure and Summary Statistics
# Check the structure of the dataset
print(lung_cancer_data.info())

# Get summary statistics
print(lung_cancer_data.describe(include='all'))
```

```python
#Handle Missing Values
# Check for missing values
missing_values = lung_cancer_data.isnull().sum()
print(missing_values)
```

```python
#Visualize Key Features
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Plot the distribution of a key numerical feature (example: age)
plt.figure(figsize=(10, 6))
sns.histplot(lung_cancer_data['AGE'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of a key categorical feature (example: gender)
plt.figure(figsize=(6, 4))
sns.countplot(x='GENDER', data=lung_cancer_data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Analyze Relationships
# Plot the relationship between 'diagnosis' and 'age'
plt.figure(figsize=(10, 6))
sns.boxplot(x='LUNG_CANCER', y='AGE', data=lung_cancer_data)
plt.title('LUNG CANCER by Age')
plt.xlabel('LUNG CANCER')
plt.ylabel('Age')
plt.show()

# Plot the relationship between 'diagnosis' and 'gender'
plt.figure(figsize=(6, 4))
sns.countplot(x='LUNG_CANCER', hue='GENDER', data=lung_cancer_data)
plt.title('LUNG_CANCER by Gender')
plt.xlabel('LUNG_CANCER')
plt.ylabel('Count')
plt.show()
```
