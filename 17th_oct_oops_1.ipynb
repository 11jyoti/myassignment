```python
#Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid deposit amount. Please deposit a positive amount.")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self.balance}")
            else:
                print("Insufficient funds. Withdrawal not allowed.")
        else:
            print("Invalid withdrawal amount. Please withdraw a positive amount.")

    def display_account_details(self):
        print(f"Account Number: {self.account_number}")
        print(f"Account Holder Name: {self.account_holder_name}")
        print(f"Balance: ${self.balance}")

# Example usage:
account1 = BankAccount("123456", "John Doe", 1000)
account1.display_account_details()

account1.deposit(500)
account1.withdraw(200)
account1.display_account_details()




```

```python
#Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self):
        # Assuming bonus is 10% of the annual salary
        return 0.1 * self.salary

    def display_employee_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: ${self.salary}")
        print(f"Yearly Bonus: ${self.calculate_yearly_bonus()}")

# Example usage:
employee1 = Employee("E001", "John Doe", 50000)
employee2 = Employee("E002", "Jane Smith", 60000)

employee1.display_employee_details()
print("\n")
employee2.display_employee_details()
```

```python
#Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.
class Vehicle:
    def __init__(self, vehicle_id, model, available=True):
        self.vehicle_id = vehicle_id
        self.model = model
        self.available = available

    def rent_vehicle(self):
        if self.available:
            self.available = False
            print(f"Vehicle {self.model} (ID: {self.vehicle_id}) has been rented.")
        else:
            print(f"Sorry, Vehicle {self.model} is currently not available for rent.")

    def return_vehicle(self):
        if not self.available:
            self.available = True
            print(f"Vehicle {self.model} (ID: {self.vehicle_id}) has been returned.")
        else:
            print(f"This vehicle is already available in the rental system.")

    def display_vehicle_details(self):
        print(f"Vehicle ID: {self.vehicle_id}")
        print(f"Model: {self.model}")
        print("Status: Available" if self.available else "Status: Not Available")


class VehicleRentalSystem:
    def __init__(self):
        self.vehicles = {}

    def add_vehicle(self, vehicle):
        if vehicle.vehicle_id not in self.vehicles:
            self.vehicles[vehicle.vehicle_id] = vehicle
            print(f"Vehicle {vehicle.model} (ID: {vehicle.vehicle_id}) added to the rental system.")
        else:
            print(f"Vehicle ID {vehicle.vehicle_id} already exists in the rental system.")

    def rent_vehicle(self, vehicle_id):
        if vehicle_id in self.vehicles:
            self.vehicles[vehicle_id].rent_vehicle()
        else:
            print(f"Vehicle ID {vehicle_id} not found in the rental system.")

    def return_vehicle(self, vehicle_id):
        if vehicle_id in self.vehicles:
            self.vehicles[vehicle_id].return_vehicle()
        else:
            print(f"Vehicle ID {vehicle_id} not found in the rental system.")

    def display_available_vehicles(self):
        if not self.vehicles:
            print("No vehicles available in the rental system.")
        else:
            print("Available Vehicles:")
            for vehicle in self.vehicles.values():
                if vehicle.available:
                    vehicle.display_vehicle_details()
                    print("\n")

# Example usage:
vehicle1 = Vehicle("V001", "Sedan")
vehicle2 = Vehicle("V002", "SUV")

rental_system = VehicleRentalSystem()
rental_system.add_vehicle(vehicle1)
rental_system.add_vehicle(vehicle2)
rental_system.display_available_vehicles()

rental_system.rent_vehicle("V001")
rental_system.display_available_vehicles()

rental_system.return_vehicle("V001")
rental_system.display_available_vehicles()


```

```python
#Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.available = True

    def borrow_book(self):
        if self.available:
            self.available = False
            print(f"Book '{self.title}' by {self.author} (Book ID: {self.book_id}) has been borrowed.")
        else:
            print(f"Sorry, '{self.title}' is currently not available.")

    def return_book(self):
        if not self.available:
            self.available = True
            print(f"Book '{self.title}' by {self.author} (Book ID: {self.book_id}) has been returned.")
        else:
            print(f"This book is already available in the library.")

    def display_book_details(self):
        print(f"Book ID: {self.book_id}")
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print("Status: Available" if self.available else "Status: Not Available")

class LibraryCatalog:
    def __init__(self):
        self.books = {}

    def add_book(self, book):
        if book.book_id not in self.books:
            self.books[book.book_id] = book
            print(f"Book '{book.title}' (Book ID: {book.book_id}) added to the library catalog.")
        else:
            print(f"Book ID {book.book_id} already exists in the library catalog.")

    def display_available_books(self):
        if not self.books:
            print("No books available in the library.")
        else:
            print("Available Books:")
            for book in self.books.values():
                if book.available:
                    book.display_book_details()
                    print("\n")

# Example usage:
book1 = Book("B001", "The Catcher in the Rye", "J.D. Salinger")
book2 = Book("B002", "To Kill a Mockingbird", "Harper Lee")

library = LibraryCatalog()
library.add_book(book1)
library.add_book(book2)
library.display_available_books()

book1.borrow_book()
library.display_available_books()

book1.return_book()
library.display_available_books()
```

```python
#Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def update_quantity(self, new_quantity):
        self.quantity = new_quantity
        print(f"Quantity for {self.name} (Product ID: {self.product_id}) updated to {new_quantity}.")

    def display_product_details(self):
        print(f"Product ID: {self.product_id}")
        print(f"Name: {self.name}")
        print(f"Price: ${self.price}")
        print(f"Quantity: {self.quantity}")

class InventorySystem:
    def __init__(self):
        self.products = {}

    def add_product(self, product):
        if product.product_id not in self.products:
            self.products[product.product_id] = product
            print(f"Product {product.name} (Product ID: {product.product_id}) added to the inventory.")
        else:
            print(f"Product ID {product.product_id} already exists in the inventory.")

    def update_product_quantity(self, product_id, new_quantity):
        if product_id in self.products:
            self.products[product_id].update_quantity(new_quantity)
        else:
            print(f"Product ID {product_id} not found in the inventory.")

    def display_available_products(self):
        if not self.products:
            print("No products available in the inventory.")
        else:
            print("Available Products:")
            for product in self.products.values():
                product.display_product_details()
                print("\n")

# Example usage:
product1 = Product("P001", "Laptop", 1200, 10)
product2 = Product("P002", "Smartphone", 800, 20)

inventory = InventorySystem()
inventory.add_product(product1)
inventory.add_product(product2)
inventory.display_available_products()

inventory.update_product_quantity("P001", 8)
inventory.display_available_products()
```

```python
#Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.
class Shape:
    def __init__(self, length=0, width=0, height=0):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        raise NotImplementedError("Subclasses must implement the calculate_area method.")

    def calculate_perimeter(self):
        raise NotImplementedError("Subclasses must implement the calculate_perimeter method.")

class Rectangle(Shape):
    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

class Triangle(Shape):
    def calculate_area(self):
        # Using Heron's formula for the area of a triangle
        s = (self.length + self.width + self.height) / 2
        return (s * (s - self.length) * (s - self.width) * (s - self.height)) ** 0.5

    def calculate_perimeter(self):
        return self.length + self.width + self.height

# Example usage:
rectangle = Rectangle(length=5, width=3)
print("Rectangle Area:", rectangle.calculate_area())
print("Rectangle Perimeter:", rectangle.calculate_perimeter())

triangle = Triangle(length=3, width=4, height=5)
print("\nTriangle Area:", triangle.calculate_area())
print("Triangle Perimeter:", triangle.calculate_perimeter())
```

```python
#Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)
        print(f"Grade {grade} added for {self.name} (Student ID: {self.student_id}).")

    def calculate_average_grade(self):
        if not self.grades:
            return 0  # Return 0 if no grades are available
        return sum(self.grades) / len(self.grades)

    def display_student_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        if self.grades:
            print(f"Grades: {', '.join(map(str, self.grades))}")
            print(f"Average Grade: {self.calculate_average_grade()}")
        else:
            print("No grades available for this student.")

# Example usage:
student1 = Student("S001", "John Doe")
student1.add_grade(90)
student1.add_grade(85)
student1.display_student_details()

```

```python
#Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.
from datetime import datetime

class Email:
    def __init__(self, sender, recipient, subject, body):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.body = body
        self.timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    def send_email(self):
        print(f"Email sent from {self.sender} to {self.recipient}.")
        print(f"Subject: {self.subject}")
        print(f"Body: {self.body}")
        print(f"Sent at: {self.timestamp}")

# Example usage:
email = Email("user1@example.com", "user2@example.com", "Meeting Tomorrow", "Hi, let's meet tomorrow at 10 AM.")
email.send_email()

```

```python
#Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post_content):
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        post = {"timestamp": timestamp, "content": post_content}
        self.posts.append(post)
        print(f"Post added successfully at {timestamp}.")

    def display_posts(self):
        if not self.posts:
            print("No posts available.")
        else:
            print(f"Posts by {self.username}:")
            for post in self.posts:
                print(f"{post['timestamp']} - {post['content']}")

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post['content'].lower()]
        if matching_posts:
            print(f"Posts containing '{keyword}':")
            for post in matching_posts:
                print(f"{post['timestamp']} - {post['content']}")
        else:
            print(f"No posts found containing '{keyword}'.")

# Example usage:
profile = SocialMediaProfile("user123")
profile.add_post("Hello, this is my first post!")
profile.add_post("Enjoying a sunny day.")
profile.display_posts()
profile.search_posts("sunny")
```

```python

#Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.
from datetime import datetime

class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task, due_date):
        self.tasks.append({"task": task, "due_date": due_date, "completed": False})
        print(f"Task '{task}' added. Due date: {due_date}")

    def mark_task_as_completed(self, task_name):
        for task in self.tasks:
            if task["task"] == task_name:
                task["completed"] = True
                print(f"Task '{task_name}' marked as completed.")
                return
        print(f"Task '{task_name}' not found in the ToDo list.")

    def display_pending_tasks(self):
        pending_tasks = [task for task in self.tasks if not task["completed"]]
        if pending_tasks:
            print("Pending Tasks:")
            for task in pending_tasks:
                print(f"Task: {task['task']}, Due Date: {task['due_date']}")
        else:
            print("No pending tasks.")

# Example usage:
todo_list = ToDoList()
todo_list.add_task("Complete project", "2023-01-31")
todo_list.add_task("Prepare presentation", "2023-02-10")
todo_list.display_pending_tasks()
todo_list.mark_task_as_completed("Complete project")
todo_list.display_pending_tasks()
```
